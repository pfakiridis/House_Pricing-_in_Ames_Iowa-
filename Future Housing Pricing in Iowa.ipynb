```r
#Principal component analysis requires only numerical variable so we reverse all teh variables to numeric.
#The system will assign a number to each level for the categorical variables
#The plots below show the variance covered from each PC, showing that the first PC covers most of the area
#The second plot shows after how many PCs we reach the treshold that we have set at 80 % of variance


names <- c(1:79)
train[,names] <- lapply(train[,names] , as.numeric)
#In order to the PCA to be accurate we need to center and scale the data so the variation of the values of each variable 
#is not significantly different
pc_price <- prcomp(train, center = T, scale. = T)
pc__var <- pc_price$sdev^2
pc__PEV <- pc__var / sum(pc__var)
par(mfrow=c(1,2))
plot(pc_price)
title(xlab="PC", ylab="Variances")
plot(
  cumsum(pc__PEV),
  ylim = c(0,1),
  xlab = 'PC',
  ylab = 'Cumulative PEV',
  pch = 20,
  col = 'orange'
)
abline(h = 0.8, col = 'red', lty = 'dashed')
```

```r
#Handling Outliers
#The handling of outliers depends entirely on the research question and the tehcnique that will be used.
#For example svm models are sensitive to outliers whilst random forests are not. It is important to handle the outliers in a way
#that do not add bias in the dataset.
par(mfrow=c(2,2))
boxplot(train$LotArea)
boxplot(train$X1stFlrSF)
boxplot(train$GrLivArea)
```

```r
#The boxplots above show the variables with the most outliers. A decision has been made to remove most of the outliers

#Removing outliers
train <- subset(train, LotArea < 20000)
train <- subset(train, X1stFlrSF < 2200)
train <- subset(train, GrLivArea < 2650)
```

```r
#splitting the train set to train it

set.seed(2018)
# create a 70/30 training/test set split
n_rows <- nrow(train)
# sample 70% (n_rows * 0.7) indices in the ranges 1:nrows
training_idx <- sample(n_rows, n_rows * 0.7)
# filter the data frame with the training indices (and the complement)
training <- train[training_idx,]
testing <- train[-training_idx,]
```

```r
#The below formula has been formed from the interactions that the variables had in the biplot. We do not want to add variables
#that do not contribute to the research question since the noise will be increased in the dataset.

library(randomForest)
set.seed(2018)

formula = SalePrice~ GrLivArea + Neighborhood + OverallQual + TotalBsmtSF + X1stFlrSF + X2ndFlrSF + GarageCars + BsmtFinSF1 + GarageArea + ExterQual + GarageType + YearBuilt + YearRemodAdd + FireplaceQu + TotRmsAbvGrd + GarageFinish + OverallCond + MSSubClass + FullBath + KitchenQual + LotArea

#Firstly we try the random forest

rf_titanic <- randomForest(formula, ntree = 500, importance = T, data = train)
```

```r
#The below variance plot shows which variables affect the most the price. It is observed that
#the Rating of basement finished area(BsmtFinSF1), the Above grade (ground) living area square feet(GrLivArea)
#and the Overall Quality play the most signifacnt role in forming the final house price.

varImpPlot(rf_titanic, type = 1)
```

```r
#Since we have applied regression in order to measure the performance of the model we need to get the root mean square error.
#The RMSE is for the random forest model is 21030.41

library(dplyr)

pred <- train%>%
  mutate(pred.reg.train = predict(rf_titanic))

mse <- pred %>%
  mutate(error = pred.reg.train - SalePrice,
         sq.error = error^2) %>%
  summarise(mse = mean(sq.error))
rmse <- sqrt(mse)
```

```r
#Trying a svm model where the RMSE falls signicantly to 17.6.

library(e1071)
mymodel1 <- svm(SalePrice~ MSZoning + LotArea + Neighborhood + Condition1 + LotArea + BldgType + OverallQual + OverallCond + YearBuilt + YearRemodAdd + ExterQual + BsmtExposure + TotalBsmtSF + GrLivArea  + BsmtFullBath + KitchenQual + Functional + FireplaceQu, data = train, kernel = "polynomial")
pred <- train%>%
  mutate(pred.reg.train = predict(mymodel1))

mse <- pred %>%
  mutate(error = pred.reg.train - SalePrice,
         sq.error = error^2) %>%
  summarise(mse = mean(sq.error))
rmse <- sqrt(mse)
rmse
```

```r
#Using ensemble techniques to train the model

library(caret)

ctrl_parameters <- trainControl(method = 'CV', number = 5)


titanic_gbm <- train(formula, data = train, method = "gbm", trControl = ctrl_parameters)
titanic_svm_radial <- train(formula, data = train, method = "svmRadial",trControl = ctrl_parameters)
titanic_svm_linear <- train(formula, data = train, method = "svmLinear",trControl = ctrl_parameters)
titanic_glm<- train(formula, data = train, method = "glm", trControl = ctrl_parameters)
titanic_treebag <- train(formula, data = train, method = "treebag", trControl = ctrl_parameters)

```

```r
#With the gbm tehcnique the value falls more at 16.5

pred <- train%>%
  mutate(pred.reg.train = predict(titanic_gbm))

mse <- pred %>%
  mutate(error = pred.reg.train - SalePrice,
         sq.error = error^2) %>%
  summarise(mse = mean(sq.error))
rmse <- sqrt(mse)
rmse
```
